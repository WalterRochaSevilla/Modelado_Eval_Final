```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.colors import LinearSegmentedColormap
from scipy.interpolate import RegularGridInterpolator
import matplotlib.colors as mcolors

from matplotlib import animation
from IPython.display import HTML
```

```python
# =================================================================
# 1. CONFIGURACIÓN DEL MODELO
# =================================================================
# Parámetros físicos
Lx, Ly, Lz = 10.0, 5.0, 2.0  # Dimensiones del lago (km)
nx, ny, nz = 40, 20, 15       # Resolución de la malla
dx, dy, dz = Lx/(nx-1), Ly/(ny-1), Lz/(nz-1)

# Coeficientes del modelo
D = 0.005       # Coeficiente de difusión (km²/day)
ux, uy, uz = 0.05, 0.02, 0.001  # Velocidades de corriente (km/day)
lambda_ = 0.01  # Tasa de decaimiento (1/day)
lambda_s = 0.002 # Tasa de sedimentación (1/day)

# Crear malla 3D
x = np.linspace(0, Lx, nx)
y = np.linspace(0, Ly, ny)
z = np.linspace(0, Lz, nz)
X, Y, Z = np.meshgrid(x, y, z, indexing='ij')

# Condición inicial (fuente gaussiana)
c0 = 5.0
sigma_x, sigma_y, sigma_z = 1.0, 0.7, 0.4

c = c0 * np.exp(-((X-1)**2/(2*sigma_x**2) + (Y-Ly/2)**2/(2*sigma_y**2) + (Z-Lz/2)**2/(2*sigma_z**2)))
```

```python
# =================================================================
# 2. SIMULACIÓN (Completando la celda 3)
# =================================================================
def run_simulation(c, nt=400):
    c_sim = c.copy()

    # Calcular paso de tiempo seguro (condiciones CFL y estabilidad)
    dt_diff = min(dx**2, dy**2, dz**2) / (6 * D)  # Estabilidad difusión
    dt_conv = min(dx/abs(ux), dy/abs(uy), dz/abs(uz))  # CFL convección
    dt = 0.1 * min(dt_diff, dt_conv)  # Factor seguridad
    
    for _ in range(nt):
        c_new = c_sim.copy()
        
        # Aplicar operadores en puntos interiores
        for i in range(1, nx-1):
            for j in range(1, ny-1):
                for k in range(1, nz-1):
                    # Término difusivo (Laplaciano)
                    laplacian = (
                        (c_sim[i+1, j, k] - 2*c_sim[i, j, k] + c_sim[i-1, j, k]) / dx**2 +
                        (c_sim[i, j+1, k] - 2*c_sim[i, j, k] + c_sim[i, j-1, k]) / dy**2 +
                        (c_sim[i, j, k+1] - 2*c_sim[i, j, k] + c_sim[i, j, k-1]) / dz**2
                    )
                    
                    # Término convectivo (Upwind para estabilidad)
                    conv_x = ux * (c_sim[i, j, k] - c_sim[i-1, j, k]) / dx if ux > 0 else ux * (c_sim[i+1, j, k] - c_sim[i, j, k]) / dx
                    conv_y = uy * (c_sim[i, j, k] - c_sim[i, j-1, k]) / dy if uy > 0 else uy * (c_sim[i, j+1, k] - c_sim[i, j, k]) / dy
                    conv_z = uz * (c_sim[i, j, k] - c_sim[i, j, k-1]) / dz if uz > 0 else uz * (c_sim[i, j, k+1] - c_sim[i, j, k]) / dz
                    
                    # Ecuación completa
                    dc_dt = D * laplacian - (conv_x + conv_y + conv_z) - lambda_ * c_sim[i, j, k] - lambda_s * c_sim[i, j, k]
                    
                    # Actualización explícita (Forward Euler)
                    c_new[i, j, k] = c_sim[i, j, k] + dt * dc_dt
        
        # Condiciones de frontera (Neumann: flujo cero)
        c_new[0, :, :] = c_new[1, :, :]     # Frontera x=0
        c_new[-1, :, :] = c_new[-2, :, :]   # Frontera x=Lx
        c_new[:, 0, :] = c_new[:, 1, :]     # Frontera y=0
        c_new[:, -1, :] = c_new[:, -2, :]   # Frontera y=Ly
        c_new[:, :, 0] = c_new[:, :, 1]     # Fondo (z=0)
        c_new[:, :, -1] = c_new[:, :, -2]   # Superficie (z=Lz)
        
        c_sim = c_new
    
    return c_sim
```

```python
c = run_simulation(c, nt=400)
```

```python

# =================================================================
# 4. VISUALIZACIÓN
# =================================================================
# =================================================================
# 1. CONFIGURACIÓN VISUAL MEJORADA
# =================================================================
# Paleta de colores para el agua (de transparente a azul profundo)
water_colors = [(0.27, 0.58, 0.78, 0.1), (0.1, 0.3, 0.5, 0.8)]
water_cmap = LinearSegmentedColormap.from_list('water', water_colors)

# Paleta para mercurio (de plateado a rojo tóxico con brillo metálico)
mercury_colors = [
    (0.95, 0.95, 0.97, 0.3),  # Plata transparente
    (0.9, 0.85, 0.1, 0.7),    # Oro líquido
    (0.8, 0.1, 0.1, 0.9)      # Rojo tóxico
]
mercury_cmap = LinearSegmentedColormap.from_list('mercury', mercury_colors)

# Añadir efecto metálico mediante normalización logarítmica
class MetalNormalize(mcolors.Normalize):
    def __call__(self, value, clip=None):
        result = super().__call__(value, clip)
        return np.power(result, 0.3)  # Ajuste para brillo metálico

norm = MetalNormalize(vmin=0, vmax=c.max())


fig = plt.figure(figsize=(20, 12))
fig.patch.set_facecolor('#e0f7fa')  # Fondo azul claro

# --- Gráfico 3D ---
ax1 = fig.add_subplot(231, projection='3d')
ax1.view_init(elev=40, azim=-60)
ax1.set_facecolor('#e0f7fa') 

# Malla de agua (volumen semi-transparente)
water = ax1.scatter(
    X, Y, Z, 
    c=Z.flatten(), 
    cmap=water_cmap,
    alpha=0.15,
    s=5,
    edgecolors='none',
    depthshade=True
)

# Mercurio (puntos metálicos con brillo)
scatter = ax1.scatter(
    X, Y, Z,
    c=c.flatten(),
    cmap=mercury_cmap,
    norm=norm,
    alpha=0.9,
    s=30,
    edgecolor=[(0.9,0.9,0.8,0.5)],
    depthshade=False,
    marker='o'
)

# Barra de color con estilo
cbar = fig.colorbar(scatter, ax=ax1, shrink=0.6, pad=0.1)
cbar.set_label('Concentración de Mercurio (mg/L)', fontsize=10)
cbar.ax.yaxis.set_tick_params(color='gold', labelcolor='darkred')

# Ajustes estéticos
ax1.set_title('Distribución 3D - Contaminación por Mercurio', 
             fontsize=12, fontweight='bold', pad=20)
ax1.set_xlabel('Longitud (km)', fontsize=9)
ax1.set_ylabel('Ancho (km)', fontsize=9)
ax1.set_zlabel('Profundidad (km)', fontsize=9)
ax1.grid(False)
ax1.xaxis.pane.fill = False
ax1.yaxis.pane.fill = False
ax1.zaxis.pane.fill = False

# =================================================================
# 3. CORTES TRANSVERSALES MEJORADOS
# =================================================================
def add_2d_plot(ax, data, title, xlabel, ylabel):
    """Función auxiliar para crear subplots 2D consistentes"""
    im = ax.imshow(
        data.T,  # Transponer para orientación correcta
        origin='lower',
        cmap=mercury_cmap,
        norm=norm,
        extent=[0, Lx, 0, Lz],
        aspect='auto'
    )
    ax.set_title(title, fontsize=11, pad=12)
    ax.set_xlabel(xlabel, fontsize=9)
    ax.set_ylabel(ylabel, fontsize=9)
    ax.grid(True, linestyle=':', alpha=0.5)
    return im

# --- Corte Longitudinal ---
ax2 = fig.add_subplot(232)
im2 = add_2d_plot(
    ax2, 
    c[:, ny//2, :], 
    f'Corte Longitudinal (y={Ly/2:.1f} km)', 
    'Longitud (km)', 
    'Profundidad (km)'
)

# --- Corte Transversal ---
ax3 = fig.add_subplot(233)
im3 = add_2d_plot(
    ax3,
    c[nx//2, :, :],
    f'Corte Transversal (x={Lx/2:.1f} km)',
    'Ancho (km)',
    'Profundidad (km)'
)

# --- Superficie ---
ax4 = fig.add_subplot(234)
im4 = add_2d_plot(
    ax4,
    c[:, :, -1],
    'Distribución en Superficie',
    'Longitud (km)',
    'Ancho (km)'
)

# --- Fondo ---
ax5 = fig.add_subplot(235)
im5 = add_2d_plot(
    ax5,
    c[:, :, 0],
    'Distribución en Fondo',
    'Longitud (km)',
    'Ancho (km)'
)

# Añadir barras de color para los cortes 2D
for im, ax in zip([im2, im3, im4, im5], [ax2, ax3, ax4, ax5]):
    cbar = fig.colorbar(im, ax=ax, shrink=0.9)
    cbar.set_label('mg/L', fontsize=8)
    cbar.ax.tick_params(labelsize=8)

# =================================================================
# 4. AJUSTES FINALES
# =================================================================
plt.tight_layout(pad=3.0)
plt.subplots_adjust(top=0.9)
fig.suptitle('Simulación de Contaminación por Mercurio - Río Pilcomayo', 
             fontsize=14, fontweight='bold', y=0.98)

# Mostrar figura
plt.show()
```

```python

# =================================================================
# 3. HERRAMIENTAS DE CONSULTA
# =================================================================
def obtener_contaminacion(x, y, z):
    """Interpola la concentración en un punto arbitrario usando trilineal"""
    # Encontrar índices circundantes
    i = np.searchsorted(x_vec, x) - 1
    j = np.searchsorted(y_vec, y) - 1
    k = np.searchsorted(z_vec, z) - 1
    
    # Coeficientes de interpolación
    dx = (x - x_vec[i]) / (x_vec[i+1] - x_vec[i])
    dy = (y - y_vec[j]) / (y_vec[j+1] - y_vec[j])
    dz = (z - z_vec[k]) / (z_vec[k+1] - z_vec[k])
    
    # Interpolación trilineal
    c00 = c[i, j, k] * (1-dx) + c[i+1, j, k] * dx
    c01 = c[i, j, k+1] * (1-dx) + c[i+1, j, k+1] * dx
    c10 = c[i, j+1, k] * (1-dx) + c[i+1, j+1, k] * dx
    c11 = c[i, j+1, k+1] * (1-dx) + c[i+1, j+1, k+1] * dx
    
    c0 = c00 * (1-dy) + c10 * dy
    c1 = c01 * (1-dy) + c11 * dy
    
    return c0 * (1-dz) + c1 * dz
x_vec = np.linspace(0, Lx, nx)
y_vec = np.linspace(0, Ly, ny)
z_vec = np.linspace(0, Lz, nz)
```

```python
concentracion = obtener_contaminacion(3.0, 2.5, 0.5)
print(f"📌 Concentración en (3.0, 2.5, 0.5): {concentracion:.4f} mg/L")
```

```python
# =================================================================
# 5. EXPERIMENTACIÓN
# =================================================================
print("\n🔬 Experimentación: Simular con el doble de difusión")

# Guardar parámetro original y CONDICIÓN INICIAL
D_original = D
c_original = c.copy()  # Guardar el array 3D inicial

# Ejecutar con nuevo parámetro
D = 2 * D_original
c_doubleD = run_simulation(c_original, nt=400)

# Calcular concentración en el mismo punto (3.0, 2.5, 0.5)
# Guardar el estado global actual de 'c'
c_original_global = c.copy()

# Temporalmente establecer 'c' a c_doubleD para que la función de interpolación lo use
c = c_doubleD
conc_doubleD_val = obtener_contaminacion(3.0, 2.5, 0.5)

# Restaurar el valor global de 'c'
c = c_original_global

# Mostrar resultados (Celda 9)
print(f"• Concentración original: {concentracion:.4f} mg/L")
print(f"• Con mayor difusión: {conc_doubleD_val:.4f} mg/L")
print(f"• Cambio: {conc_doubleD_val - concentracion:.4f} mg/L")

# Restaurar valor original de D
D = D_original
```

```python
print("\n📊 TABLAS DE PROBABILIDAD CONDICIONAL (CPDs):")

# Importar la clase correcta
from pgmpy.models import DiscreteBayesianNetwork 
from pgmpy.factors.discrete import TabularCPD

model = DiscreteBayesianNetwork([ 
    ('Fuente_Industrial', 'Contaminacion'),
    ('Velocidad_Corriente', 'Contaminacion'),
    ('Profundidad', 'Contaminacion'),
    ('Contaminacion', 'Riesgo_Humano')
])
# Definir CPDs
cpd_fuente = TabularCPD(
    variable='Fuente_Industrial',
    variable_card=2,
    values=[[0.3], [0.7]],  # 30% Alta, 70% Baja
    state_names={'Fuente_Industrial': ['Alta', 'Baja']}
)

cpd_velocidad = TabularCPD(
    variable='Velocidad_Corriente',
    variable_card=2,
    values=[[0.6], [0.4]],  # 60% Lenta, 40% Rápida
    state_names={'Velocidad_Corriente': ['Lenta', 'Rápida']}
)

cpd_profundidad = TabularCPD(
    variable='Profundidad',
    variable_card=2,
    values=[[0.5], [0.5]],  # 50% Fondo, 50% Superficie
    state_names={'Profundidad': ['Fondo', 'Superficie']}
)

# CPD para Contaminación (depende de los 3 factores)
cpd_contaminacion = TabularCPD(
    variable='Contaminacion',
    variable_card=3,  # Baja, Media, Alta
    values=[
        # (Fuente, Velocidad, Profundidad)
        [0.8, 0.6, 0.7, 0.5, 0.4, 0.3, 0.2, 0.1],  # Baja
        [0.15, 0.3, 0.2, 0.3, 0.4, 0.4, 0.3, 0.2], # Media
        [0.05, 0.1, 0.1, 0.2, 0.2, 0.3, 0.5, 0.7]  # Alta
    ],
    evidence=['Fuente_Industrial', 'Velocidad_Corriente', 'Profundidad'],
    evidence_card=[2, 2, 2],
    state_names={
        'Contaminacion': ['Baja', 'Media', 'Alta'],
        'Fuente_Industrial': ['Alta', 'Baja'],
        'Velocidad_Corriente': ['Lenta', 'Rápida'],
        'Profundidad': ['Fondo', 'Superficie']
    }
)

# CPD para Riesgo Humano
cpd_riesgo = TabularCPD(
    variable='Riesgo_Humano',
    variable_card=2,
    values=[
        [0.9, 0.7, 0.3],  # Seguro cuando Baja, Media, Alta
        [0.1, 0.3, 0.7]   # Peligroso cuando Baja, Media, Alta
    ],
    evidence=['Contaminacion'],
    evidence_card=[3],
    state_names={
        'Riesgo_Humano': ['Seguro', 'Peligroso'],
        'Contaminacion': ['Baja', 'Media', 'Alta']
    }
)

# Añadir CPDs al modelo
model.add_cpds(cpd_fuente, cpd_velocidad, cpd_profundidad, cpd_contaminacion, cpd_riesgo)

# Verificar modelo
model.check_model()

# Mostrar CPDs (Celda 11)
for cpd in model.get_cpds():
    print(f"\n{cpd}")
```
